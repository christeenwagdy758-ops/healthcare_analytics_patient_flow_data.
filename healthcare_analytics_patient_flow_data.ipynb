```python
import pandas as pd
import numpy as np
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hassanjameelahmed/healthcare-analytics-patient-flow-data")

print("Path to dataset files:", path)
```

```python
import os

os.listdir(path)
```

```python
df = pd.read_csv(path + '/healthcare_analytics_patient_flow_data.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Create boolean column — True if patient was referred by a department, False if they came on their own
df['has_referral'] = df['Department Referral'].notna()
```

```python
# Create boolean column — True if patient submitted a satisfaction score, False if they didn't
df['has_ Satisfaction'] = df['Patient Satisfaction Score'].notna()
```

```python
df
```

```python
# Count patients by referral status — True: referred from a department, False: walked in alone
df['has_referral'].value_counts()
```

```python
# Department that referred the most patients to the ER
df['Department Referral'].value_counts()
```

```python
# Calculate average patient wait time in minutes
df['Patient Waittime'].mean()
```

```python
Patient_Referral_wait=df.groupby('Department Referral')['Patient Waittime'].mean()
Patient_Referral_wait
```

```python
Patient_Admission_flag_wait=df.groupby('Patient Admission Flag')['Patient Waittime'].mean()
Patient_Admission_flag_wait
```

```python
conditions = [
    df['Patient Age'] <= 18,
    df['Patient Age'] <= 35,
    df['Patient Age'] <=60 ,
    df['Patient Age'] >60
]

choices = ['Children', 'Young Adults', 'Adults', 'Seniors']

df['Age'] = np.select(conditions, choices, default='Seniors')
```

```python
df['Age'].value_counts()
```

```python
df['Patient Gender'].value_counts()
```

```python
df['Patient Gender'] = df['Patient Gender'].replace('Femaleemale', 'Female')
```

```python
df['Patient Gender'].value_counts()
```

```python
df['Patient Race'].value_counts()
```

```python
df['Patient Admission Time'].dtype
```

```python
df['Patient Admission Date']=pd.to_datetime(df['Patient Admission Date'],dayfirst=True)
```

```python
df['Month']=df['Patient Admission Date'].dt.month
df['Month']
```

```python
monthly_Patient =df.groupby('Month')['Patient Id'].count()
monthly_Patient
```

```python
monthly_Patient .idxmax()
```

```python
visit_counts=df['Patient Id'].value_counts()
returning = visit_counts[visit_counts > 1]
returning
```
